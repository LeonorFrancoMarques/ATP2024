```python
def menu():
    
    print("---------- MENU ----------")
    print("Opção 1 - Criar uma turma")
    print("Opção 2 - Inserir um aluno na turma")
    print("Opção 3 - Listar a turma")
    print("Opção 4 - Consultar um aluno por id")
    print("Opção 5 - Guardar a turma em ficheiro")
    print("Opção 6 - Carregar uma turma dum ficheiro")
    print("Opção 0 - Sair da aplicação")

menu()

import random

def criarAluno():
    nome = input("Nome do aluno: ")
    id = int(input("ID do aluno: "))
    notaTPC = float(input("Nota do TPC: "))
    notaProj = float(input("Nota do projeto: "))
    notaTeste = float(input("Nota do teste:" ))
    return(nome, id, [notaTPC, notaProj, notaTeste])

def criarTurma():
    turma = []
    return turma

def inserirAluno(turma):
    aluno = criarAluno()
    if aluno not in turma:
        turma.append(aluno)
        print("O aluno foi inserido na turma!")
    else:
        print("Esse aluno já está na turma!")
    return turma

def listarTurma(turma):
    print("---------- Turma ----------")
    for aluno in turma:
        print(f"Nome: {aluno[0]}, ID: {aluno[1]}, Notas: TPC: {aluno[2][0]}, Projeto: {aluno[2][1]}, Teste: {aluno[2][2]}")
    return

def idAluno(turma, id):
    for aluno in turma:
        if aluno[1] == id:
            print(f"Nome: {aluno[0]}, ID: {aluno[1]}, Notas: TPC: {aluno[2][0]}, Projeto: {aluno[2][1]}, Teste: {aluno[2][2]}")
    return

def guardarTurma(turma, fnome):
    file = open(fnome, "w")
    for aluno in turma:
        nome, id, notas = aluno[0], aluno[1], aluno[2]
        notaTPC, notaProj, notaTeste = notas
        for nota in notas:
            n = (str(notaTPC) + " & " + str(notaProj) + " & " + str(notaTeste))
        file.write(f"{nome} :: {id} :: {n}\n")
    file.close()

def carregarTurma(fnome):
    turma = []
    file = open(fnome, "r")
    for line in file:
        parcelas = line.split(" :: ")
        nome, id, notas = parcelas[0], parcelas[1], parcelas[2]
        nota = notas.split(" & ")
        notaTPC, notaProj, notaTeste = nota[0], nota[1], nota[2]
        aluno = (nome, int(id), [float(notaTPC), float(notaProj), float(notaTeste)])
        turma.append(aluno)
    file.close()
    return turma


opcao = 1
turma = []
cond = True
while opcao != 0 and cond:
    opcao = int(input("Introduza a sua opção:"))
    if opcao == 1:
        criarTurma()
        print("A turma foi criada!")
    elif opcao == 2:
        inserirAluno(turma)
    elif opcao == 3:
        listarTurma(turma)
    elif opcao == 4:
        id = int(input("Introduza o ID do aluno que quer consultar:"))
        idAluno(turma, id)
    elif opcao == 5:
        guardarTurma(turma, "turma.txt")
    elif opcao == 6:
        turma_a = carregarTurma("turma.txt")
        print(turma_a)
    else:
        if opcao != 0:
            print("Resposta inválida!")
        elif opcao == 0:
            cond = False
if opcao == 0:
    print("Aplicação encerrada, volte sempre!")


turma1 = []

print(turma1)
print(listarTurma(turma1))
```
